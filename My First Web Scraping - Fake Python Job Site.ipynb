```python
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://realpython.github.io/fake-jobs/')
```

```python
type(page)
```

```python
print(page.text)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
type(soup)
```

```python
results = soup.find(id = 'ResultsContainer')
print(results.prettify())
```

```python
job_elements = results.find_all('div', class_='card-content')
```

```python
for job_element in job_elements:
    print(job_element, end='\n'*2)
```

```python
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    print(title_element)
    print(company_element)
    print(location_element)
    print()
```

```python
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    print(title_element.text)
    print(company_element.text)
    print(location_element.text)
    print()
```

```python

for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    print(title_element.text.strip())
    print(company_element.text.strip())
    print(location_element.text.strip())
    print()
```

```python
Data = []
element_list = []
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    element_list.append(title_element.text.strip())
    element_list.append(company_element.text.strip())
    element_list.append(location_element.text.strip())
    Data.append(element_list)
    element_list = []
    
print(Data)
```

```python
Data = []
element_list = {}
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    element_list.update(Title = title_element.text.strip())
    element_list.update(Company = company_element.text.strip())
    element_list.update(Location = location_element.text.strip())
    Data.append(element_list)
    element_list = {}
    
print(Data)
```
